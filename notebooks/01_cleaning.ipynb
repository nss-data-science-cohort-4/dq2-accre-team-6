```python
def node_lists(value):
    if '[' not in value:
        return [value]
    
    nodes = []
    node_index = re.findall(r'([a-z-]+)|$', value)[0]
    for item in re.findall(r'(\[.*\])', value)[0].replace('[','').replace(']','').split(','):
        if '-' in item:
            top, bottom = re.findall(r'(\d+)', item)
            nodes.extend([f'{node_index}{x}' for x in range(int(top), int(bottom)+1)])
        else:
            nodes.append(f'{node_index}{item}')
    return nodes
```

```python
accre_jobs_piped['NODELIST'] = accre_jobs_piped['NODELIST'].apply(node_lists)
```

```python
def get_seconds(used_time_string):
    """calculate total second in job time"""
    days = 0
    day_break = used_time_string.split('-')
    if len(day_break) > 1:
        days = int(day_break[0])
        hms = day_break[1]
    else:
        hms = day_break[0]
    hours, mins, secs = [int(part) for part in hms.split(':')]
    total_seconds = secs + (mins * 60) + (hours * 3600) + (days * 86400)
    return total_seconds    
```

```python
accre_jobs_piped['req_total_seconds'] = accre_jobs_piped['reqtime'].apply(get_seconds)
```

```python
def get_reqmb_per_core(row):
    if 'Mn' in row['reqmem']:
        cpus = row['cpus'] or 1
        reqmem = int(row['reqmem'][:-2]) / (cpus / row['nodes'])
    elif 'Mc' in row['reqmem']:
        reqmem = int(row['reqmem'][:-2])
    else:
        reqmem = row['reqmem']
    return reqmem  
```

```python
accre_jobs_piped['reqmem_mc'] = accre_jobs_piped.apply(get_reqmb_per_core, axis = 1)
```

```python
def get_usemb_per_core(row):
    cpus = row['cpus'] or 1
    if 'M' in row['usedmem']:
        reqmem = float(row['usedmem'][:-1]) / (cpus / row['nodes'])
    else:
        reqmem = float(row['usedmem']) / (cpus / row['nodes'])
    return reqmem  
```

```python
accre_jobs_piped['usedmem_mc'] = accre_jobs_piped.apply(get_usemb_per_core, axis = 1)
```
